```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')

path_crime = "/content/drive/My Drive/Data Files/RQ1.csv"
#path_acs = "/content/drive/My Drive/Data Files/acs_tracts_travis_2019.csv"

rq_1 = pd.read_csv(path_crime)
#acs_data = pd.read_csv(path_acs)
```

```python
print(rq_1.shape)
print(rq_1.isnull().sum())
```

```python
cols_dropped = ['incident_number', 'offense_code', 'apd_sector', 'apd_district', 'council_district', 'census_block_group', 'clearance_date']

rq_1_df = rq_1.drop(cols_dropped, axis=1)

print(rq_1_df.shape)
print(rq_1_df.isnull().sum())
```

```python
rq_1_df = rq_1_df.dropna()
print(rq_1_df.shape)
print(rq_1_df.isnull().sum())
```

```python
rq_1_df['occurred_datetime'] = pd.to_datetime(rq_1_df['occurred_datetime'])
rq_1_df['report_datetime'] = pd.to_datetime(rq_1_df['report_datetime'])
```

```python
rq_1_2010s = rq_1_df[
    (rq_1_df['occurred_datetime'].dt.year >= 2016) &
    (rq_1_df['occurred_datetime'].dt.year <= 2023)
].copy()

rq_1_2010s.describe(include='all')
```

```python
str_cols = ['tract_geoid']
for col in str_cols:
    rq_1_2010s[col] = rq_1_2010s[col].apply(lambda x: format(int(x), 'd'))

print(rq_1_2010s[str_cols].dtypes)
rq_1_2010s.head(10)
```

```python
value_counts_offense = rq_1_2010s['offense'].value_counts()
#print(value_counts_offense)

common_offenses = value_counts_offense[value_counts_offense > 500]
print(common_offenses)

print("Sum of offenses: ", common_offenses.sum())

print("Sum of offenses, not adjusted: ", value_counts_offense.sum())
```

```python
rq_1_df_common = rq_1_2010s[rq_1_2010s['offense'].isin(common_offenses.index)]
rq_1_df_common.shape
```

```python
tract_counts = rq_1_df_common['tract_geoid'].value_counts()
print(tract_counts.sum())

common_tract = tract_counts[tract_counts > 400]
print(common_tract)
print(common_tract.sum())
```

```python
rq_1_df_common2 = rq_1_df_common[rq_1_df_common['tract_geoid'].isin(common_tract.index)]
rq_1_df_common2.shape
print(rq_1_df_common2.head(5))
rq_1_df_common2.describe(include='all')
```

```python
rq_1_df_common2.loc[:,'family_violence'] = rq_1_df_common2['family_violence'].map({'Y': 1, 'N': 0})
print(rq_1_df_common2.head(5))
```

```python
value_counts_offense2 = rq_1_df_common2['offense'].value_counts()
print(value_counts_offense2)
print(value_counts_offense2.sum())

tract_counts2 = rq_1_df_common2['tract_geoid'].value_counts()
print(tract_counts2)
print(tract_counts2.sum())

family_violence_counts = rq_1_df_common2['family_violence'].value_counts()
print(family_violence_counts)
```

```python
rq_1_df_common2.describe(include='all')
```

```python
rq_1_df_common2.shape
```

```python
offense_to_category = {
    'THEFT BY SHOPLIFTING': 'PROPERTY',
    'DWI': 'PUBLIC_ORDER',
    'WARRANT ARREST NON TRAFFIC': 'ADMINISTRATIVE',
    'THEFT': 'PROPERTY',
    'POSSESSION OF MARIJUANA': 'DRUG',
    'CUSTODY ARREST TRAFFIC WARR': 'ADMINISTRATIVE',
    'POSS CONTROLLED SUB/NARCOTIC': 'DRUG',
    'POSS OF DRUG PARAPHERNALIA': 'DRUG',
    'ASSAULT W/INJURY-FAM/DATE VIOL': 'VIOLENT',
    'CRIMINAL TRESPASS': 'PUBLIC_ORDER',
    'VIOL CITY ORDINANCE - OTHER': 'PUBLIC_ORDER',
    'ASSAULT WITH INJURY': 'VIOLENT',
    'AGG ASSAULT': 'VIOLENT',
    'UNLAWFUL CARRYING WEAPON': 'PUBLIC_ORDER',
    'DWI 2ND': 'PUBLIC_ORDER',
    'DWI .15 BAC OR ABOVE': 'PUBLIC_ORDER',
    'DEL CONTROLLED SUB/NARCOTIC': 'DRUG',
    'FAILURE TO IDENTIFY': 'ADMINISTRATIVE',
    'AGG ASSAULT': 'VIOLENT',
    'VOCO ALCOHOL CONSUMPTION': 'PUBLIC_ORDER',
    'VIOL STATE LAW - OTHER': 'PUBLIC_ORDER',
    'THEFT OF SERVICE': 'PROPERTY',
    'AUTO THEFT': 'PROPERTY',
    'CRIMINAL MISCHIEF': 'PROPERTY',
    'FELONY ENHANCEMENT/ASSLT W/INJ': 'VIOLENT',
    'DRIVING WHILE INTOX / FELONY': 'PUBLIC_ORDER',
    'EVADING / FOOT': 'PUBLIC_ORDER',
    'FAMILY DISTURBANCE': 'PUBLIC_ORDER',
    'BURGLARY OF VEHICLE': 'PROPERTY',
    'BURGLARY NON RESIDENCE': 'PROPERTY',
    'AGG ROBBERY/DEADLY WEAPON': 'VIOLENT',
    'AGG ASSAULT FAM/DATE VIOLENCE': 'VIOLENT',
    'BURGLARY OF RESIDENCE': 'PROPERTY',
    'EVADING / FOOT' : 'PUBLIC_ORDER',
}

rq_1_df_common3 = rq_1_df_common2.copy()
rq_1_df_common3['offense'] = rq_1_df_common3['offense'].map(offense_to_category).fillna('OTHER')

rq_1_df_common3.shape

rq_1_df_common3.describe(include='all')
```

```python
category_offense_counts = rq_1_df_common3['offense'].value_counts()
print(category_offense_counts)
print(category_offense_counts.sum())
```

```python
tract_geoid_counts = rq_1_df_common3['tract_geoid'].value_counts()
print(tract_geoid_counts)
```

```python
save_path = "/content/drive/My Drive/Data Files/cleaned_rq1.csv"

rq_1_df_common3.to_csv(save_path, index=False)
```
